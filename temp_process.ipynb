```python
reader = open('organ.csv', 'r', encoding='gbk')
lines = reader.readlines()
lines = [o.strip('\n').split(',')[0] for o in lines]
high_school_count = 0
school_count = 0
research_count = 0

for line in lines:
    if '大学' or '学院' in line:
        high_school_count += 1
    elif '学校' in line:
        school_count += 1
    else:
        research_count += 1

print(high_school_count, school_count, research_count)
```

```python
import pymongo
import itertools

session = pymongo.MongoClient().get_database('cnki_db').get_collection('非遗数字化传承')
items = list(session.find())
funds_items = list(map(lambda o: o['funds'], items))
funds = list(itertools.chain(*funds_items))
print(funds)
for fund in funds:
    print(fund)
    
```
